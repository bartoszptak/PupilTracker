```
import cv2
import matplotlib.pyplot as plt
import numpy as np
data_set = np.load('arrayU.npy')

scale = 2.99

print(data_set[0][0].shape, data_set[0][1], data_set[0][2])

for im in data_set:
  im[0] = cv2.resize(im[0],None, None, scale, scale)
  im[1] = int(im[1]*scale)
  im[2] = int(im[2]*scale)

plt.imshow(data_set[0][0])
print(data_set[0][0].shape, data_set[0][1], data_set[0][2])
```

```
X_ = data_set[:,0:1]
Y_ = data_set[:,1:3]

Xx = []
Y = []

for x in X_:
    Xx.append(x[0] * 1. /255)
    
Xx = np.array(Xx)    
Y = np.array(Y_)

Y *= 1. / (100 * scale)

X = Xx

print(len(X),X.shape)
print(len(Y),Y.shape)
```

```
from keras.models import Sequential
from keras.models import Model
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, EarlyStopping, ReduceLROnPlateau, TensorBoard
from keras import optimizers, losses, activations, models
from keras.layers import Convolution2D, Dense, Input, Flatten, Dropout, MaxPooling2D, BatchNormalization, GlobalAveragePooling2D, Concatenate
from keras import applications

input_shape = (int(scale*100),int(scale*100), 3)
nclass = 2

base_model = applications.InceptionV3(weights='imagenet', include_top=False, input_shape=input_shape)
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(2, activation='linear')(x)

model = Model(inputs=base_model.input, outputs=predictions)

for layer in model.layers[:249]:
   layer.trainable = False
for layer in model.layers[249:]:
   layer.trainable = True

model.compile(loss='mean_absolute_error', optimizer=optimizers.SGD(lr=1e-4,momentum=0.9), metrics=['mean_absolute_error'])
model.summary()
```

```
model.fit(X,Y,epochs=500, shuffle=True, verbose=True, validation_split=0.3)
```

```
model.save('fajnyTenModelMaszSzmaciuro2.h5')
model.save_weights('zacneWagiMilordzie2.h5')
```
